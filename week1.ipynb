```python

import pandas as pd
import matplotlib.pyplot as plt
import ast

md = pd.read_csv('tmdb_5000_movies.csv')
# md.dropna(inplace=True)
genres = {}
count = 0
for i in md.index:
    temp = ast.literal_eval(md.loc[i,'genres'])
    for x in temp:
        if (x["name"] in genres):
            genres[x["name"]] += 1
            count += 1
        else:
            genres[x["name"]] = 1
            count += 1

plt.figure(figsize=(12, 6)) 
plt.bar(list(genres.keys()),list(genres.values()))
plt.xlabel("Genres")
plt.xticks(rotation=60)
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()
print(count)
# print((md.to_numpy()))
```
